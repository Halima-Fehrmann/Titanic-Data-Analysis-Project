# Create a function and apply to the grouped DataFrame

- We will define a function called 'mean_and_round' which passes two arguments: a value (x) and value a (a=2).
- Take 'x' (in each row) and calcaluate the mean of 'x' using numpy functions.
- Round the mean of 'x' by the value of 'a', which will be passed in to the argument. 
- Finally, return the mean value.

```python
def mean_and_round(x, a=2):
    mean = np.round(np.mean(x, axis=0), a)
    return mean
```

```python
# 'fare' & 'age' values are rounded to 1 decimal place, using the custom function, and assigned to a new DataFrame.  

rounded_fare_age = df.groupby(['embarked', 'pclass'])[
    ['fare', 'age']].apply(mean_and_round, 1)

rounded_fare_age
```

# We can apply custom functions to groups of dataframes, which will return a DataFrame

We will define a function named **'top_n_females'**.  This will return a subset of the titanic dataset with the top paying female passengers. The function takes two arguments; a DataFrame and a value (n) in the specified order.

- Filter the DataFrame by the columns that include only female passengers. 
- Sort the values by fare, in descending order.
- Assign this to a new DataFrame called **'female_subset'**, this will form the first argument in the function
- The second argument creates a DataFrame based on the first DataFrame. The second DataFrame returns all the row values using index positions starting at 0 and upto but not including position n. This is assigned to a new DataFrame named **'top_n_subset'**.

```python

def top_n_females(df, n=5):
    
    female_subset = df.loc[df['sex'] == 'female'].sort_values(
        'fare', ascending=False)
    
    top_n_subset = female_subset[0:n]

    return top_n_subset
```

```python
# top 3 highest paying passengers all of whom survived
top_n_females(df, n=3)
```

# The top 4 highest paying females, who survived, were all first class passengers.

```python
# Group the entire DataFrame by the column 'survived' and apply the function to extract the 4 top paying females. 

top_females_by_survival = df.groupby('survived').apply(top_n_females, n=4)
top_females_by_survival
```

# .agg () and Lambda functions

We will use .agg() on a groupby object to apply functions to different columns. For this, we will create a dictionary with key-value pairs. 

We will also include lambda functions within the dictionary.  

Below we have created our own function and instead of passing .agg() a dictionary, we have passed it a list that will apply every function in that list to each column, in the groupby output.

```python
# convert according to made-up conversion, then find mean

def money_conversion_mean(x):
    converted = x/12 * 5
    return np.mean(converted)


df.groupby('pclass').agg({'survived': 'mean',
                               'age': ['mean', 'median', 'count', 'std'],
                               'sibsp': [lambda x: np.mean(x)],
                               'fare': ['mean', money_conversion_mean]})
```

```python
# group by 'sex' and 'pclass', specify the columns and apply functions to each column using .agg()

grouped_agg = df.groupby(['sex', 'pclass'])[
    ['survived', 'age', 'fare']].agg(['mean', 'median', 'std'])
grouped_agg
```

# Apply functions to compare columns 

```python
# compare 'pclass' and 'class' to see if they are giving the same information 
# we will use a dictionary to compare the rows in multiple columns 

def check_class_cols(row):
    
    class_val = row["pclass"]
    class_text = row["class"]

    class_dict = {1: 'First', 2: 'Second', 3: 'Third'}
    
    return class_dict[class_val] == class_text   
```

```python
# apply the function to each row

titanic['class_check'] = titanic[['pclass', 'class']].apply(check_class_cols, axis=1)
```

```python
# count the number of true values. true value indicates the compared values are equal

titanic['class_check'].sum() == len(titanic)
```

```python
titanic.head()
```

```python
# The values in 'pclass' and 'class' are equal - we can delete one of these columns 

titanic['class_check'].unique()
```
